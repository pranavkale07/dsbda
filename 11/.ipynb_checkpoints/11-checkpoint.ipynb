```python
import pandas as pd
```

```python
df=pd.read_csv('movies_metadata.csv', low_memory=False)

display(df.head())
display(df.columns)
```

```python
# a. Create data subsets for different languages(Original Language).  

english_subset=df[df['original_language']=='en']
tamil_subset=df[df['original_language']=='ta']

display(english_subset.head(5))
display(tamil_subset.head(5))
```

```python
# b. Merge two subsets

merged_data=pd.concat([english_subset, tamil_subset])
display(merged_data.head(5))
```

```python
display(df.dtypes)
```

```python
# c. Sort Data using customer ratings. (vote_average)

sorted_data=df.sort_values(by='vote_average',ascending=False)

display(sorted_data[['title', 'vote_average', 'vote_count']].head(10))
display(sorted_data.head(5))
```

```python
# d. Transposing Data 

transposed_data=df.transpose()

display(transposed_data)
```

```python
# e. Melting Data to long format

melted_data=pd.melt(df,id_vars=['original_title'],value_vars=['vote_average','vote_count'])

display(melted_data)
```

```python
# f. Casting data to wide format 

casted_data=melted_data.pivot_table(index='original_title',columns='variable',values='value',aggfunc='sum').reset_index()

display(casted_data)
```
