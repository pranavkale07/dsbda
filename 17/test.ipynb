```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('heart.csv')
print(df.head(5))
df.columns
```

```python
# a. histogram

plt.figure(figsize=(10,4))
plt.hist(df['age'],bins=10,edgecolor='black')
plt.title('Age Vs Count')
plt.xlabel('Age')
plt.ylabel('Count')
```

```python
# b. Dotplots

plt.figure(figsize=(10,4))
plt.plot(df['age'],df['chol'],'o',color='purple')
plt.title('Age vs Cholesterol')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
```

```python
# c. barplots

plt.figure(figsize=(10,4))
sns.barplot(df,x='age',y='chol',palette='Set1')
plt.title('Age vs Cholesterol')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
```

```python
# d. Line Charts

sorted=df.sort_values(by='age')
plt.figure(figsize=(10,4))
plt.plot(sorted['age'],sorted['chol'],marker='o')
plt.title('Age vs Cholesterol')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
```

```python
# e. pie chart

counts=df['age'].value_counts()

plt.pie(counts,labels=counts.index)
```
