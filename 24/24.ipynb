```python
import pandas as pd
from IPython.display import display
```

```python
df=pd.read_csv('forestfires.csv')

display(df)
```

```python
df.dtypes
```

```python
# a. Create data subsets by making classes for amount of region affected.(e.g. NotAffected, Partially affected, Mostlyaffected). 

def classify_area(area):
    if area == 0:
        return 'NotAffected'
    elif area <= 30:
        return 'PartiallyAffected'
    else:
        return 'MostlyAffected'
    
df['RegionalClass']=df['area'].apply(classify_area)

NotAffected_subset=df[df['RegionalClass']=='NotAffected']
PartiallyAffected_subset=df[df['RegionalClass']=='PartiallyAffected']
MostlyAffected_subset=df[df['RegionalClass']=='MostlyAffected']


display(NotAffected_subset.head())
display(PartiallyAffected_subset.head())
display(MostlyAffected_subset.head())
```

```python
# b. Merge two subsets 

merged_data=pd.concat([NotAffected_subset,MostlyAffected_subset])
display(merged_data)
```

```python
# c. Sort Data using Temperature, wind and area. 

sorted_temp=df.sort_values(by='temp',ascending=False)           #separate

sorted_data=df.sort_values(by=['temp','wind','area'],ascending=False)       #together

sorted_data.head()
```

```python
# d. Transposing

transposed_data=df.head(4).T

transposed_data
```

```python
# e. Melting Data to long format

melted_data= pd.melt(df,id_vars=['month'],value_vars=['temp','wind','rain'])

display(melted_data)
```

```python
# f. Casting data to wide format  

casted_data= melted_data.pivot_table(index='month',columns='variable',values='value',aggfunc='sum').reset_index()

casted_data
```
