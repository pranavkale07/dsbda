```python
import pandas as pd
from IPython.display import display
# Define the correct column names as per dataset description
column_names = [
    "Class", "Age", "Sex", "Steroid", "Antivirals", "Fatigue", "Malaise", "Anorexia",
    "Liver_Big", "Liver_Firm", "Spleen_Palpable", "Spiders", "Ascites", "Varices",
    "Bilirubin", "Alk_Phosphate", "SGOT", "Albumin", "Protime", "Histology"
]

# Load the Hepatitis dataset
df = pd.read_csv('hepatitis.csv', header=None, names=column_names)

# Display first few rows
display(df)
```

```python
# Subset for males (assuming 'male' is encoded as 'M' or 1)
male_df = df[df['Sex'] == 1]  # or == 1 depending on encoding
female_df = df[df['Sex'] == 2]  # or == 2

print("Male Patients:")
display(male_df.head())

print("Female Patients:")
display(female_df.head())
```

```python
# Merge male and female data back together
merged_df = pd.concat([male_df, female_df], ignore_index=True)

print("Merged Data:")
display(merged_df)
```

```python
# Convert SGOT and PROTIME to numeric (handle "?")
df["SGOT"] = pd.to_numeric(df["SGOT"], errors="coerce")
df["Protime"] = pd.to_numeric(df["Protime"], errors="coerce")

# Now sort
sorted_df = df.sort_values(by=["Age", "SGOT", "Protime"])
print("\nSorted Data:")
display(sorted_df[["Age", "SGOT", "Protime"]])
```

```python
transposed = df.transpose()

print("Transposed Data:")
display(transposed)
```

```python
# Example: melt 5 rows on id_vars 'Class' and 'Age'
melted_df = pd.melt(df.head(), id_vars=["Class", "Age"])
print("\nMelted Data (long format):")
display(melted_df)

```

```python
# Pivot back (wide) using melted data
cast_df = melted_df.pivot(index=["Class", "Age"], columns="variable", values="value").reset_index()
print("\nRe-casted Data (wide format):")
display(cast_df)

```
