```python
import pandas as pd

# Load dataset
df = pd.read_csv('dataset_Facebook.csv', delimiter=';')

# a. Create data subsets for type of post
photo_posts = df[df['Type'] == 'Photo']
status_posts = df[df['Type'] == 'Status']
display("Subset: Photo Posts\n", photo_posts.head())
display("Subset: Status Posts\n", status_posts.head())

# b. Merge two subsets (e.g., concatenating photo and status posts)
merged_posts = pd.concat([photo_posts, status_posts], ignore_index=True)
display("Merged Subset\n", merged_posts.head())

# c. Sort Data on Page total likes
sorted_df = df.sort_values(by='Page total likes', ascending=False)
display("Sorted by Page Total Likes\n", sorted_df[['Page total likes', 'Type']].head())

# d. Transposing Data (flip rows and columns of a small subset)
transposed = df.head(3).transpose()
display("Transposed Data\n", transposed)

# e. Melting Data to long format
# Selecting a few columns to melt for demonstration
melted = pd.melt(df,
                 id_vars=['Type', 'Post Month'],
                 value_vars=['Page total likes', 'Lifetime Post Total Reach'],
                 var_name='Metric',
                 value_name='Value')
display("Melted Data\n", melted.head())

# f. Casting data to wide format (pivoting back from melted form)
casted = melted.pivot_table(index=['Type', 'Post Month'],
                            columns='Metric',
                            values='Value',
                            aggfunc='mean').reset_index()
display("Casted Data (Wide Format)\n", casted.head())
```
