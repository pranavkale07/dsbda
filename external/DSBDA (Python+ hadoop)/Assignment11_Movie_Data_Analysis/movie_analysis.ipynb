# Movie Data Analysis

```python
import pandas as pd

file_path = "../datasets/Movie/movies_metadata.csv"
df = pd.read_csv(file_path, delimiter=",", low_memory=False)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['original_language'].unique()
```

```python
# Create data subsets for different languages(Original Language). 
en_df = df[df['original_language'] == 'en']
print("Count of en movies : ", len(en_df))

hi_df = df[df['original_language'] == 'hi']
print("Count of hi movies : ", len(hi_df))

for lang in df['original_language'].unique():
    lang_df = df[df['original_language'] == lang]
    print(f"Count of '{lang}' movies: {len(lang_df)}")
```

```python
# Merge two subsets 
merged_df = pd.concat([en_df, hi_df], ignore_index=True)
merged_df.head()
```

```python
# Sort Data using customer ratings.  
sorted_df = df.sort_values('vote_average', ascending=False)
sorted_df[['title', 'vote_average']].head()
```

```python
# Transposing Data
transposed_df = df.transpose()
transposed_df
```

```python
# Melting Data to long format 
melted_df = pd.melt(df, id_vars=['title', 'original_language'], value_vars=['budget', 'revenue', 'runtime'], var_name='Metric', value_name='Value')
melted_df
```

```python
# Casting data to wide format  (long to wide)
casted_df = melted_df.pivot_table(values='Value', index=['title', 'original_language'], columns='Metric', aggfunc='first').reset_index()
casted_df
```
