# Hepatitis Data Analysis

```python
import pandas as pd

file_path = "../datasets/Hepatitis/hepatitis_csv.csv"
df = pd.read_csv(file_path, delimiter = ",")
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.info()
```

```python
# Create data subsets for different sex.  
male_df = df[df['sex'] == 'male']
female_df = df[df['sex'] == 'female']

print("Count of male : ", len(male_df))
print("Count of female : ", len(female_df))

male_df
```

```python
# Merge two subsets 
merged_df = pd.concat([male_df, female_df], ignore_index=True)
merged_df.head()
```

```python
# Sort Data using age, SGOT, PROTIME
sorted_df = df.sort_values(['age', 'sgot', 'protime'], ascending=True)
sorted_df[['sex','age', 'sgot', 'protime']]
```

```python
# Transposing Data 
transposed_df = df.transpose()
transposed_df
```

```python
# Melting Data to long format (Wide -> Long)
melt_df = pd.melt(df, id_vars=['age', 'sex', 'histology', 'class'], value_vars=['steroid', 'antivirals', 'fatigue', 'malaise', 'anorexia', 'liver_big', 'liver_firm', 'spleen_palpable', 'spiders', 'ascites', 'varices'], var_name = 'Symptom', value_name='Presence')
melt_df
```

```python
# Casting data to wide format (Long -> Wide)
cast_df = melt_df.pivot_table(index='Symptom', columns='class', values='Presence', aggfunc='mean').reset_index()
cast_df
```
