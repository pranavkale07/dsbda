```python
# melt(): Transforms data from wide format to long format. You gather many columns into one.
# pivot(): Transforms data from long format to wide format. You spread out one column into many.


melted_df = pd.melt(df,id_vars=['Type'],value_vars=['comment', 'share','like'],var_name='Engagement_Type',value_name='Engagement_Value')
# id_vars:Columns to keep as identifiers (these columns will remain as they are).
# value_vars:Columns that you want to melt into a single column.
# var_name:The name of the new column created from the melted column headers (variables).
# value_name:The name of the new column that holds the values.

melted_df
```

```python
wide_df = melted_df.pivot_table(index=['Type'],columns='Engagement_Type',values='Engagement_Value',aggfunc='mean').reset_index()
print(wide_df)

# ðŸ“Œ index=['Type']
# This tells Pandas to group rows by the Type column.
# Each unique value in 'Type' (e.g., Photo, Status, Link, Video) will become a row in the output

# ðŸ“Œ columns='Engagement_Type'
# This determines the new column names in the wide-format table.
# Values in the 'Engagement_Type' column (e.g., 'like', 'share', 'comment', etc.) will become separate columns in the new DataFrame.


# ðŸ“Œ values='Engagement_Value'
# This is the actual data you want to fill into the table.
# Values from the 'Engagement_Value' column (numeric values like 45 likes, 10 comments) are placed into the corresponding cells.

# ðŸ“Œ aggfunc='mean'
# If there are multiple rows with the same Type and Engagement_Type, this tells Pandas to take the average (mean) of their values.
# You can change this to 'sum', 'max', 'min', etc., depending on what kind of aggregation you want.
```

```python
us_subset = df[df['native-country'] == 'United-States']
male_subset = df[df['gender'] == 'Male']
white_subset = df[df['race'] == 'White']

merged_df = pd.concat([us_subset, male_subset, white_subset], ignore_index=True)
sorted_df = merged_df.sort_values(by='hours-per-week', ascending=False)
transposed_df = sorted_df.T
melted_df = pd.melt(sorted_df, id_vars=['age', 'income'], value_vars=['workclass', 'education', 'capital-gain', 'hours-per-week'])
pivot_table_df = melted_df.pivot_table(index='age',columns='variable',values='value',aggfunc='first').reset_index()     
```

```python
english_movies = df[df['original_language'] == 'en']
french_movies = df[df['original_language'] == 'fr']

merged_movies = pd.concat([english_movies, french_movies], ignore_index=True)
sorted_movies = merged_movies.sort_values(by='vote_average', ascending=False)
transposed_movies = sorted_movies.T
melted_movies = pd.melt(sorted_movies, id_vars=['original_title', 'original_language'], value_vars=['vote_average', 'genres'])
pivoted_movies = melted_movies.pivot_table(index='original_title', columns='original_language', values='vote_average', aggfunc='first')
```

```python
def classify_area(area):
    if area == 0:
        return 'NotAffected'
    elif 0 < area <= 10:
        return 'PartiallyAffected'
    else:
        return 'MostlyAffected'

df['area_class'] = df['area'].apply(classify_area)

partially_affected = df[df['area_class'] == 'PartiallyAffected']
mostly_affected = df[df['area_class'] == 'MostlyAffected']

merged_df = pd.concat([partially_affected, mostly_affected], ignore_index=True)
sorted_df = df.sort_values(by=['temp', 'wind', 'area'], ascending=False)
transposed_df = df.T

melted_df = pd.melt(df, id_vars=['X', 'Y', 'month', 'day'], value_vars=['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain', 'area'])
pivoted_df = df.pivot_table(index='X', columns='month', values='temp', aggfunc='mean')
```
