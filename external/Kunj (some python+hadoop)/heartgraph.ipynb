```python
import pandas as pd
import numpy as np

cleavl_read=pd.read_csv('./DSBDALExam DataSets/HeartDisease/Cleavland.csv',header=None)
hung_read=pd.read_csv('./DSBDALExam DataSets/HeartDisease/hung.csv',header=None)
switzer_read=pd.read_csv('./DSBDALExam DataSets/HeartDisease/Switzerland.csv',header=None)

column_names = [
    "age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach",
    "exang", "oldpeak", "slope", "ca", "thal", "target"
]
cleavl_read.columns=column_names
hung_read.columns=column_names
switzer_read.columns=column_names
print("Cleveland Data Set", cleavl_read.head(5))
print("Hungarian Data Set", hung_read.head(5))
print("Switzerland Data Set", switzer_read.head(5))
```

```python
df=pd.concat([cleavl_read,hung_read,switzer_read],ignore_index=True)
df
```

```python
df.replace('?',np.nan,inplace=True)
df.dropna(inplace=True)
df = df.apply(pd.to_numeric)
df.dtypes
```

```python
for col in df.columns:
    df= df[df[col]>=0]

df
```
