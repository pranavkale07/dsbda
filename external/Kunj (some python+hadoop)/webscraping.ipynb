```python
import requests
import bs4
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'Cache-Control': 'no-cache',
    'Pragma': 'no-cache'
}


url = "https://www.flipkart.com/poco-c31-royal-blue-64-gb/p/itm19effae969b86"

request1 = requests.get(url, headers=headers)
request1.content
```

```python
request1.content
```

```python

soup =bs4.BeautifulSoup(request1.content,'html.parser')
print(soup.prettify())
```

```python
reviews=soup.find_all('div',{'class':'ZmyHeo'})
for review in reviews:
    print(review.text)
    print('------------------')
```

```python
rating=soup.find_all('div',class_='XQDdHH Ga3i8K')
for review in rating:
    print(review.text)
    print('------------------')
```

```python
names=soup.find_all('p',class_='z9E0IG')
for name in names:
    print(name.text)
    print('------------------')
```

```python
commetTags=soup.findAll('div',{'class':'wys2hv _43gOsC'})
for commetTag in commetTags:
    print(commetTag.get_text())
    answer=commetTag.findNext('div',{'class':'JxAXcP'}).get_text()
    print(answer)
    print('------------------')
```

```python
import bs4
import requests
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'Cache-Control': 'no-cache',
    'Pragma': 'no-cache'
}
request1=requests.get('https://www.imdb.com/title/tt31193180/reviews/?ref_=tt_ururv_sm',headers=headers)
```

```python
request1.content
soup=bs4.BeautifulSoup(request1.content)
soup
```

```python
revies=soup.find_all('div',class_='ipc-html-content ipc-html-content--base')
for review in revies:
    print(review.text)
    print('------------------')
```

```python
ratings=soup.find_all('span',class_='ipc-rating-star--rating')
for rating in ratings:
    print(rating.get_text())
    print('------------------')
```
